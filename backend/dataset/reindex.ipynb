```python
import json
import numpy as np
import scipy as scs
import pandas as pd
from pandas.io.json import json_normalize 

print("Starting...")

data = json.load(open('formatted_tweets.json'))
data = data["tweets"]
json_normalize(data)
df = pd.DataFrame(data)

print("starting...")

list_id = sorted(set(df["id"]))
df["id"] = df["id"].apply(lambda x: list_id.index(x))

print("done...")

list_id = sorted(set(df["user_id"]))
df["user_id"] = df["user_id"].apply(lambda x: list_id.index(x))
```

```python
print("done...")


df.to_csv("cleaned_dataset", index=False)


print(df.head())
```
